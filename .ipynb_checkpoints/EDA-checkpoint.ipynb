#### 缺失值统计

```python
# Missing Value Function
def missing_data(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum()/data.isnull().count()*100)
    tt = pd.concat([total, percent], axis=1, keys=['Total','Percent'])
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    tt['Types'] = types
    return(np.transpose(tt))
```

```python
missing_data(train_df)
```

```python
missing_data(test_df)
```

可以看到训练集和测试集都没有缺失值，于是我们又省力了，不用担心缺失值的问题了，但是一般来说有缺失值的话，可以按照以下步骤进行：
- 若特征缺失比例超过阈值，则可以直接删除特征，这个阈值一般在60%-99%之间，与数据集本身相关
- 若非第一种情况，则一般有以下的办法：
    - 插均值
    - 线性插值
    - 插众数值
    - 。。。

视特征情况而定。

而杀手锏则是用随机森林进行缺失值补全。

### 进一步的数据统计描述

#### 训练集和测试集的某些特征的散点图

正如1.3.3所示，训练集和测试集的数据特征几乎一模一样，所以我们可以检查下对应特征的散点图，假如特征的每一行真的非常接近的话，那么它应该近似为一条直线$Y = X$，但是对于20万行的数据，这几乎不可能，但是可以肯定的是，假如我们的猜想正确，那么数据的散点图应该是一个非常靠近中心的圆圈状的图，且"离圈点"应该很少。

```python
def plot_scatter(train_df,test_df,n):
    features = test_df.columns[1:n]
    plt.figure()
    fig, ax = plt.subplots(4,math.ceil(n/4),figsize=(15,15))
    
    i = 0
    for feature in features:
        i += 1
        plt.subplot(4,math.ceil(n/4),i)
        plt.scatter(train_df[feature],test_df[feature],marker= "+")
        plt.xlabel(feature)
    plt.show()
```

```python
plot_scatter(train_df,test_df,17)
```

可以看到，和我们猜测的差不多，训练集和测试集的数据很接近。

#### 相同Target的数据特征的分布情况

正如上面所说，数据集处于失衡状态，那么假如不同label下的数据特征分布又是近似的话，那么预测会更加的难搞，先看看密度图

```python
def plot_density(df1,df2,label1,label2,features):
    plt.figure()
    fig,ax = plt.subplots(10,10,figsize=(18,22))
    i = 0
    
    for feature in features:
        i += 1
        plt.subplot(10,10,i)
        sns.kdeplot(df1[feature],label = label1)
        sns.kdeplot(df2[feature],label = label2)
        plt.xlabel(feature)
        locs, labels = plt.xticks()
        plt.tick_params(axis='x', which='major', labelsize=6, pad=-6)
        plt.tick_params(axis='y', which='major', labelsize=6)
    plt.show()
```

```python
df1 = train_df.loc[train_df['target'] == 0]
df2 = train_df.loc[train_df['target'] == 1]
features = test_df.columns[1:101]
plot_density(df1,df2,'0','1',features)
```

```python
features = test_df.columns[101:]
plot_density(df1,df2,'0','1',features)
```

可以看到，绝大部分特征都是一样的，这些特征对于分类是没有什么太大帮助的，我们要找的是密度曲线不一样的特征。

大部分特征密度函数呈正态分布密度函数形态，但是有少部分呈高斯混合分布形态，比如典型的var_2.

这些特征以后也是要多考虑的，因为我们知道现实世界中分布大多呈单一的正态分布，这种特征几乎是司空见惯的了，
对分类没什么太大帮助

#### 训练集和测试集特征的分布情况

```python
features = test_df.columns[1:101]
plot_density(train_df,test_df,'train','test',features)
```

```python
features = test_df.columns[101:]
plot_density(train_df,test_df,'train','test',features)
```

可以发现，训练集和测试集的数据特征近乎一致，训练的模型一般能很好的预测测试集，因为我们最害怕的就是测试集和训练集不平衡，
比如极端情况下，训练集给出的全是label为1的数据，而测试集实际是label为0的数据，可以想见这种情况下，密度图肯定是不一样的，这样训练集
训练出来的东西几乎无法很好的预测测试集。

#### 统计每个特征中的重复值个数

```python
features = train_df.columns[2:202]
train_max_feature = []
test_max_feature = []
for feature in features:
    values = train_df[feature].value_counts()
    train_max_feature.append([feature,values.max(),values.idxmax()])
    values = test_df[feature].value_counts()
    test_max_feature.append([feature,values.max(),values.idxmax()])
```

```python
np.transpose((pd.DataFrame(train_max_feature, columns=['Feature', 'Max duplicates', 'Value'])).\
            sort_values(by = 'Max duplicates', ascending=False).head(15))
```

```python
np.transpose((pd.DataFrame(test_max_feature, columns=['Feature', 'Max duplicates', 'Value'])).\
            sort_values(by = 'Max duplicates', ascending=False).head(15))
```

可以看到，训练集和测试集上的重复值较多的feature是基本一致的，这样就给我们减少了很多麻烦，因为有时候会遇到训练集和测试集不一致的情况，这样就很难取舍。
